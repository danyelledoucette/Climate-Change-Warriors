```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import sem
from scipy.stats import linregress

path1="../../Resources/tas_1901_2016_-43.196388_-22.90833_RioBrazil.csv"
path2="../../Resources/tas_1901_2016_106.816666_-6.2_JakartaIndonesia.csv"
path3="../../Resources/tas_1901_2016_-90.071533_29.951065_NewOrleansUS.csv"
path4="../../Resources/tas_1901_2016_-117.853104_33.787914_OrangeCountyUS.csv"

rRio=pd.read_csv(path1)
rJakarta=pd.read_csv(path2)
rNewOrleans=pd.read_csv(path3)
rOrangeCounty=pd.read_csv(path4)

import gmaps
from config import gkey
gmaps.configure(api_key=gkey)
```

```python
rRio["City"]="Rio de Janeiro"
rRio["Country"]="Brazil"
rRio=rRio.rename(columns={" Year" : "Year", " Statistics" : "Statistics", " Longitude" : "Longitude", " Latitude" : "Latitude"
                    })
rRio.head()
```

```python
#plt.figure(figsize=(20,10))
plt.plot(rRio["Year"],rRio["Temperature - (Celsius)"])
plt.ylabel("Temperature - (Celsius)")
plt.xlabel("Year")
plt.show()
```

```python
rRio_yrave=rRio.groupby(["Year"]).mean()
rRio_yrave["Temperature - (Celsius)"]
rRio["Year"][0]
rRio["Year"].max()
rRio_yrave["Temperature - (Celsius)"]
rRio_yr=pd.Series(range(rRio["Year"][0], rRio["Year"].max()+1))
plt.plot(rRio_yr, rRio_yrave["Temperature - (Celsius)"])
plt.title("Rio de Janeiro, Brazil")
plt.ylabel("Temperature - (Celsius)")
plt.xlabel("Year")
plt.show()
```

```python
(slope, intercept, rvalue_rRio, pvalue, stderr) = linregress(rRio_yr, rRio_yrave["Temperature - (Celsius)"])
rvalue_Rio1=round(rvalue_rRio,2)
regress_values = rRio_yr * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x " + str(round(intercept,2))
plt.scatter(rRio_yr,rRio_yrave["Temperature - (Celsius)"])
plt.plot(rRio_yr,regress_values,"r-")
plt.annotate(line_eq,(1980,21.5),fontsize=10,color="red")
plt.ylabel("Temperature - (Celsius)")
plt.xlabel("Year")
print(f"The correlation between time (year) and annual temperature is: {rvalue_Rio1}")
plt.show()
```

```python
rRio_yravedel=rRio_yrave["Temperature - (Celsius)"].diff().dropna(how='any')
x_axis=pd.Series(range(rRio_yr[1], rRio_yr.max()+1))
plt.bar(x_axis, rRio_yravedel, color='g', alpha=1, align='center')
plt.ylabel("Temperature Difference from Previous Year (Celsius)")
plt.xlabel("Year")
plt.title("Rio de Janeiro")
plt.show()
```

```python
rRio1=pd.DataFrame(rRio_yrave["Temperature - (Celsius)"]) 
rJakarta1=pd.DataFrame(rJakarta_yrave["Temperature - (Celsius)"]) 
rNewOrleans1=pd.DataFrame(rNewOrleans_yrave["Temperature - (Celsius)"]) 
rOrangeCounty1=pd.DataFrame(rOrangeCounty_yrave["Temperature - (Celsius)"])
```

```python
for b in city_list:
    if b == "Rio de Janeiro, Brazil":
        quartiles=rRio1.quantile([.25,.5,.75])
        ave=rRio1.mean()
    elif b == "Jakarta, Indonesia":
        quartiles=rJakarta1.quantile([.25,.5,.75])
        ave=rJakarta1.mean()
    elif b=='New Orleans, USA':
        quartiles=rNewOrleans1.quantile([.25,.5,.75])
        ave=rNewOrleans1.mean()
    elif b=='Orange County, USA':
        quartiles=rOrangeCounty1.quantile([.25,.5,.75])    
        ave=rOrangeCounty1.mean()
        
    Q1=quartiles["Temperature - (Celsius)"][0.25]
    Q3=quartiles["Temperature - (Celsius)"][0.75]
    iqr=(quartiles["Temperature - (Celsius)"][0.75]-quartiles["Temperature - (Celsius)"][0.25])
    B1=Q1 - (1.5*iqr)
    B2=quartiles["Temperature - (Celsius)"][0.5] 
    B3=Q3 + (1.5*iqr)    
    
    if b == "Rio de Janeiro, Brazil":
        IQ_rRio=rRio1.describe().transpose()
        IQ_rRio["IQR"]=iqr
        IQ_rRio["Lower Bound"]=B1
        IQ_rRio["Upper Bound"]=B3        
        IQ_rRio["City"]="Rio de Janeiro"
        
    elif b == "Jakarta, Indonesia":
        IQ_rJakarta=rJakarta1.describe().transpose()
        IQ_rJakarta["IQR"]=iqr
        IQ_rJakarta["Lower Bound"]=B1
        IQ_rJakarta["Upper Bound"]=B3  
        IQ_rJakarta["City"]="Jakarta"    
        
    elif b=='New Orleans, USA':
        IQ_rNewOrleans=rNewOrleans1.describe().transpose()
        IQ_rNewOrleans["IQR"]=iqr
        IQ_rNewOrleans["Lower Bound"]=B1
        IQ_rNewOrleans["Upper Bound"]=B3 
        IQ_rNewOrleans["City"]="New Orleans"    
        
    elif b=='Orange County, USA':
        IQ_rOrangeCounty=rOrangeCounty1.describe().transpose()
        IQ_rOrangeCounty["IQR"]=iqr
        IQ_rOrangeCounty["Lower Bound"]=B1
        IQ_rOrangeCounty["Upper Bound"]=B3         
        IQ_rOrangeCounty["City"]="Orange County" 
```

```python
frames_IQ=[IQ_rRio, IQ_rJakarta, IQ_rNewOrleans, IQ_rOrangeCounty]
combine_IQ=pd.concat(frames_IQ)
combine_IQ=combine_IQ.rename(columns={"count" : "Count", "mean" : "Mean_Temperature",
                                      "std" : "Std_Dev_Temperature", "min" : "Minimum_Temperature",
                                      "25%" : "Q1", "50%" : "Median_Temperate",
                                      "75%" : "Q3", "max" : "Maximum_Temperature"
                                 })
combine_IQ.columns
```

```python
combine_IQn=combine_IQ[['City', 'Count', 'Mean_Temperature', 'Std_Dev_Temperature', 'Minimum_Temperature', 'Maximum_Temperature', 
                           'Q1','Q3', 'IQR', 'Lower Bound','Upper Bound', 'Median_Temperate']]
combine_IQn1=combine_IQn.set_index('City').transpose()
combine_IQn1
```
